```python
# Python example: quicksort
from random import choice,sample

def quicksort(l):
    if len(l)<2: return l
    pivot = choice(l)
    lo = [x for x in l if x<pivot]
    eq = [x for x in l if x==pivot]
    hi = [x for x in l if x>pivot]
    return quicksort(lo)+eq+quicksort(hi)

```
