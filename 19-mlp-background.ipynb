# Multilayer Perceptrons

(multi-layer networks)

So far...

- individual neurons can implement logic gates
- we can group together neurons in layers
- each layer consists of gates that compute in parallel
- can perform arbitrary computations with single layer recurrent networks
- can perform linear classification (perceptron learning, logistic regression)

Disadvantages of recurrent networks:

- potentially unbounded latency
- hard to program
- hard to apply machine learning to

(multi-layer networks)

Solution:

- stack multiple perceptron layers on top of each other

What boolean functions can we compute with a fixed number of layers and polynomial units? 

- circuit complexity
- common complexity class of log-depth, polynomial-sized circuits

What decision regions can we compute with a fixed number of layers and polynomial units? 

- machine learning

(circuit complexity in TOC)

- layers of gates (AND/OR/NOT/NAND...)
- each layer operates fully parallel
- complexity is given by depth (=delay), number of gates, and fan-in/fan-out
- for each input size $N$, we construct a circuit sufficient for solving problems of size $N$

(complexity of the parity function)

$\pi\not\in AC^0$

$AC^0$ = polynomial size, unlimited fan-in, AND+OR, depth $O(1)$, NOT gates at input

$\pi\in NC^1$

$NC^1$ = polynomial size, polylogarithmic depth, AND/OR/NOT gates

(what happens in real neural networks?)

- fixed latency in many computational tasks $\rightarrow$ feed forward operations

- variable latency in many $\rightarrow$ recurrent operations

What can we do with feed-forward networks?

(Linearly Nonseparable Problem)

Many problems are not linearly separable. These cannot be solved with a single layer of perceptrons. To solve these, we need _multi-layer perceptrons_.

(Multilayer Perceptrons)

Current multilayer perceptrons:

$$ y = \phi(A\cdot x + a) $$
$$ z = \phi(B\cdot y + b) $$

Note that this can be converted into a single-layer recurrent network:

$$ y = \phi(M \cdot {x \choose 0} + m) $$
$$ {0 \choose z} = \phi(M \cdot y + m) $$

(Complexity Theory for Learning)

- large number of results on "learnability" and "complexity of learning problems"
- theory has been used to justify algorithms like support vector machines
- learning algorithms are applied without actually checking the conditions for learnability anyway
